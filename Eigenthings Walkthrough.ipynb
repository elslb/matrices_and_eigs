# Definitions
___

* Eigenvectors are vectors that only change in scale when a matrix is applied to them.:
    * i.e. They keep the same direction but only change in "size"




```python
import numpy as np
import matplotlib.pyplot as plt

# This sets the default style for the figures
plt.style.use('fivethirtyeight')
```

```python
# Create a transformation matrix

A = np.array([[-1, 3], [2, -2]])
A
```

```python
# Create a vector

v = np.array([[2], [1]])
v
```

```python
def plotVectors(vecs, colors, alpha=1, figsize=(10,8)):
    """
    Plot set of vectors.

    Parameters
    ----------
    vecs : array-like
        Coordinates of the vectors to plot. Each vectors is in an array. For
        instance: [[1, 3], [2, 2]] can be used to plot 2 vectors.
    colors : array-like
        Colors of the vectors. For instance: ['red', 'blue'] will display the
        first vector in red and the second in blue.
    alpha : float
        Opacity of vectors
    figsize : tuple (x_size,y_size)
        Size of the figure

    Returns:

    fig : instance of matplotlib.figure.Figure
        The figure of the vectors
    """
    plt.figure(figsize=figsize)
    plt.axvline(x=0, color='#A9A9A9', zorder=0)
    plt.axhline(y=0, color='#A9A9A9', zorder=0)

    vecs = [vec.flatten() for vec in vecs]

    for i in range(len(vecs)):
        x = np.concatenate([[0,0],vecs[i]])
        plt.quiver([x[0]],
                   [x[1]],
                   [x[2]],
                   [x[3]],
                   angles='xy', scale_units='xy', scale=1, color=colors[i],
                   alpha=alpha)
```

```python
# Getting the dot product of the matrix and the vector
Av = A.dot(v)
Av
```

```python
# Visualizing the vectors on a plot

plotVectors([v, Av], colors = ['red', 'blue'])
plt.ylim(-1, 4)
plt.xlim(-1, 4)
plt.show()
```

```python
A = np.array([[5, 1], [3, 3]])
A
```

```python
# This is the eigenvector of the new A array
v1 = np.array([[1], [1]])
v1
```

```python
Av1 = A.dot(v1)
Av1
```

## Note about eig()
___
numpy linalg module has a function eig() returns eigenvectors and eigenvalues
 * The first array contains the eigenvalues (not ordered by magnitude)
 * The second contains the eigenvectors concatenated into columns
___

```python
eigenvals, V = np.linalg.eig(A)
```

```python
eigenvals, V
```

## Now we're reconstructing Matrix A from its eigenvectors and values
___

```python
# creating a diagonal matrix from eigenvalues

lambdas = np.diag(eigenvals)
lambdas
```

```python
# computing the inverse matrix of the eigenvector matrix "V"

V_inv = np.linalg.inv(V)
V_inv
```

___
#### Reconstruct A from the dot product of the the following:
* eigenvector matrix
* diagonal matrix made from the eigenvalues
* inverse of the eigenvector matrix
___

```python
A_orig = V.dot(lambdas).dot(V_inv)
A_orig
```
